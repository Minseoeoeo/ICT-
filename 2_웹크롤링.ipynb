####requests 로 html (하이퍼텍스 마크업 랭귀지) 가져오기

```python
import requests
# requests는 웹사이트에 접속해서 데이터를 받아오거나 보낼 수 있도록 도와주는 도구(라이브러리)라는 뜻

url = "https://news.naver.com/"
# http 인터넷 통신 규약, https
# p: protocol 규정, 규약, s: secure 보안
response = requests.get(url)
response
# <Response [200]> 200: 정상 수신 프로토콜
```

```python
response.text
```

```python
html = response.text
print(html[:500])  # 앞에서부터 500개만 가져와 -> (slicing)
```

#### Beautiful Soup 으로 html 파싱하다
https://pypi.org/project/beautifulsoup4/

```python
# 파싱은 분해한다는 말
from bs4 import BeautifulSoup
# 뷰티풀숩: 네이버 뉴스 사이트 접근해서 너가 파싱해줘. 그걸 숩이라고 해두고, 이후에 숩에서
soup = BeautifulSoup(html, "html.parser")

# 주요 뉴스 제목 가져오기 (CSS Selector)
# headlines = soup.select("a[href^='https://news.naver.com/section/101']")
headlines = soup.select("a[href^='https://n.news.naver.com']")

# print(headlines)

# h 는 그냥 인덱스라서 i로 적어도 상관없는거임
for h in headlines[:5]:
    print(h.text)

# print(h) 라고 하면 전체 코드를 가져오는데, h.text라고 하면 text만 가져옴
```

```python
# 파싱은 분해한다는 말
from bs4 import BeautifulSoup
# 뷰티풀숩: 네이버 뉴스 사이트 접근해서 너가 파싱해줘. 그걸 숩이라고 해두고, 이후에 숩에서
soup = BeautifulSoup(html, "html.parser")

# 주요 뉴스 제목 가져오기 (CSS Selector)
# headlines = soup.select("a[href^='https://news.naver.com/section/101']")
headlines = soup.select("a[href^='https://n.news.naver.com']")

# print(headlines)

for index in headlines[:5]:
    print(index.text.strip())
    # index: 리스트에서 순번 (0번, 1번, 2번, 3번, 4번: 5개 있네)
    # text: 한글이나 영어처럼 문자만 추추
    # strip() : 양쪽 공백 제거

#자료수집 끝!
```

#####가져온 데이터를 판다스 (pandas) 정리

```python
import pandas as pd
# as : alias 별칭(별명)
# >> 앞으로 pandas 를 pd 라고 부를게

data = []  #초기화

for head in headlines[:10]:
    title = head.text.strip()
    link = head['href']    # href 는 'HTML' 에서 링크를 지정할 때 쓰는 속성임
    data.append({"제목":title, "링크":link})

data
```

####가져온 데이터를 판다스로 정리

```python
import pandas as pd

pd.DataFrame(data)
# 리스트 >> 데이터 프레임 형태 (엑셀같은 행과 열로 구성되어 있음)
```

```python
df = pd.DataFrame(data)
df.index = range(1, len(df)+1)
df
```

#### 텍스트 전처리: 불필요한 문자 제거

```python
import re
# regular expression: 정규표현식

def clean_text(text):
    text = re.sub(r'[^\w\s]', '', text)  # special word 특수문자 제거 후 공백으로 둬서 text로 업데이트
    text = re.sub(r'\s+', ' ', text)     # space 공백으로 시작하는 애들 정리
    return text.strip()                  # 양쪽 공백 제거

# 위와 같은 정규표현식은 그냥 외우지 말고 가져다 쓰면 됨. 그리고 gpt 한테 물어보면 돼.
# gpt에게 목적 말하고 이런걸 위한 함수 짜달라고!
```

```python
df_loaded['제목']
```

```python
df_loaded['제목'].apply(clean_text)
# apply: (함수를) 적용하다
```

```python
df_loaded["제목_정제"] = df_loaded['제목'].apply(clean_text)
df_loaded
```
